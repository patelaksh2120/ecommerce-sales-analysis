# Data Preprocessing

## A. Data Cleaning

````python
# Typecasting: Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Optional: Confirm data types after typecasting
print(df.dtypes)
````

````python
# Step 1: Calculate missing value percentages
missing_percent = df.isnull().mean() * 100

# Step 2: Define actions based on rules
drop_rows_cols = missing_percent[missing_percent < 15].index.tolist()
impute_cols = missing_percent[(missing_percent >= 20) & (missing_percent <= 70)].index.tolist()
drop_columns = missing_percent[missing_percent > 75].index.tolist()

# Step 3: Drop rows where <15% columns have nulls
df_cleaned = df.dropna(subset=drop_rows_cols)

# Step 4: Impute missing values for 20%‚Äì70% columns
for col in impute_cols:
    if df_cleaned[col].dtype in ['float64', 'int64']:
        df_cleaned[col].fillna(df_cleaned[col].median(), inplace=True)
    else:
        df_cleaned[col].fillna(df_cleaned[col].mode()[0], inplace=True)

# Step 5: Drop columns with >75% missing data
df_cleaned.drop(columns=drop_columns, inplace=True)

# Final check
print(df_cleaned.info())
````

````python
# Assuming 'df' is your DataFrame and it has 'Month_Name' and 'Total Revenue' columns.
# For demonstration purposes, let's create a dummy DataFrame that resembles the expected data structure.
# In your actual notebook, you would already have 'df' loaded.
data = {
    'Month_Name': ['January', 'February', 'March', 'April', 'May', 'June',
                   'July', 'August', 'September', 'October', 'November', 'December'] * 2,
    'Total Revenue': [
        10000, 11000, 12000, 15000, 18000, 22000,
        25000, 28000, 32000, 45000, 50000, 48000,
        10500, 11500, 12500, 15500, 18500, 22500,
        25500, 28500, 32500, 45500, 50500, 48500 # Slightly different values for a second "year"
    ]
}
df = pd.DataFrame(data)

# --- Enhanced Code Start ---

# 1. Group by 'Month_Name' and calculate the sum of 'Total Revenue'
# Ensure the months are in chronological order for plotting
month_order = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]
monthly_sales = df.groupby('Month_Name')['Total Revenue'].sum().reindex(month_order)

# 2. Create the plot using matplotlib and seaborn for better aesthetics
plt.figure(figsize=(12, 7)) # Adjust figure size for better readability

# Use sns.barplot directly or plt.bar. Using plt.bar for direct value labeling.
bars = plt.bar(monthly_sales.index, monthly_sales.values, color=sns.color_palette("viridis", 12)) # Use a color palette

plt.title('Total Revenue by Month', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Total Revenue (INR)', fontsize=14) # Specify units if known, e.g., INR, USD

plt.xticks(rotation=45, ha='right', fontsize=12) # Rotate labels and align them
plt.yticks(fontsize=12)

# Add value labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + (yval * 0.01), round(yval, 0), ha='center', va='bottom', fontsize=9)
    # Adjust yval * 0.01 to position text slightly above the bar

plt.grid(axis='y', linestyle='--', alpha=0.7) # Add a horizontal grid for easier reading of values
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
````

## Average Revenue per Transaction by Payment Method

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Create dummy data with Payment Method and Total Revenue columns
np.random.seed(42)
payment_methods = ['Credit Card', 'Debit Card', 'Digital Wallet', 'Bank Transfer', 'Cash']
data = {
    'Payment Method': np.random.choice(payment_methods, 500),
    'Total Revenue': np.random.uniform(10, 500, 500)
}
df = pd.DataFrame(data)

# Calculate average revenue by payment method
avg_revenue_payment = df.groupby('Payment Method')['Total Revenue'].mean().sort_values()

# Create the visualization
plt.figure(figsize=(10, 6))

# Gradient colors from light to dark blue
colors = plt.cm.Blues_r(np.linspace(0.3, 1, len(avg_revenue_payment)))

bars = plt.barh(
    avg_revenue_payment.index,
    avg_revenue_payment.values,
    color=colors,
    edgecolor='black',
    linewidth=0.7
)

# Add value labels
for bar in bars:
    width = bar.get_width()
    plt.text(
        width + (0.02 * max(avg_revenue_payment)),
        bar.get_y() + bar.get_height()/2,
        f'${width:,.2f}',
        va='center',
        fontsize=10
    )

plt.title('Average Transaction Value by Payment Method', 
         fontsize=15, pad=20, fontweight='bold')
plt.xlabel('Average Revenue per Transaction (USD)', fontsize=12, labelpad=10)
plt.ylabel('Payment Method', fontsize=12, labelpad=10)
plt.xlim(0, max(avg_revenue_payment) * 1.15)

# Add reference line for overall average
overall_avg = df['Total Revenue'].mean()
plt.axvline(overall_avg, color='red', linestyle='--', alpha=0.7)
plt.text(
    overall_avg * 1.01, 
    len(avg_revenue_payment)-0.5, 
    f'Overall Avg: ${overall_avg:,.2f}',
    color='red'
)

plt.grid(axis='x', alpha=0.3)
sns.despine(left=True)
plt.tight_layout()
plt.show()
````
